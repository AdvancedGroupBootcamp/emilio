```python
# Import libraries
import requests as rq
import pandas as pd
import numpy as np
import re
```

```python
# Read API site, transform it into Json file while reading all the pages from each one of the available links.
# the data will be returned in a dataframe
base_urls = rq.get('https://swapi.dev/api/').json()
base_urls
def getAllResponse(url):
    data_response = rq.get(url).json()
    data = []
    data+=data_response['results']

    while(data_response['next'] != None):
        data_response = rq.get(data_response['next']).json()
        data+=data_response['results']
    data = pd.DataFrame(data)   
    return data
    
```

```python
# Function to clean columns and find all possible outcomes.
def cleanColumns(columns, data):
    for column in columns:
        data[column] = data[column].map(lambda x :  [re.findall(r'\d+', url)[0] for url in x])
```

```python
# Saved urls into python vairables
people = getAllResponse(base_urls['people'])
films = getAllResponse(base_urls['films'])
vehicles = getAllResponse(base_urls['vehicles'])
starships = getAllResponse(base_urls['starships'])
species = getAllResponse(base_urls['species'])
planets = getAllResponse(base_urls['planets'])
```

```python
# database creation
conn = sqlite3.connect('star_wars.db')
c = conn.cursor()

c.execute(f"CREATE TABLE IF NOT EXISTS specialities ({', '.join(films.columns)})")
conn.commit()
films.to_sql('films', conn, if_exists='replace', index = True)

c.execute(f"CREATE TABLE IF NOT EXISTS physicians ({', '.join(people.columns)})")
conn.commit()
people.to_sql('people', conn, if_exists='replace', index = True)

c.execute(f"CREATE TABLE IF NOT EXISTS consultations ({', '.join(vehicles.columns)})")
conn.commit()
vehicles.to_sql('vehicles', conn, if_exists='replace', index = True)

c.execute(f"CREATE TABLE IF NOT EXISTS patients ({', '.join(starships.columns)})")
conn.commit()
starships.to_sql('starships', conn, if_exists='replace', index = True)

c.execute(f"CREATE TABLE IF NOT EXISTS patients ({', '.join(species.columns)})")
conn.commit()
species.to_sql('species', conn, if_exists='replace', index = True)

c.execute(f"CREATE TABLE IF NOT EXISTS patients ({', '.join(planets.columns)})")
conn.commit()
planets.to_sql('planets', conn, if_exists='replace', index = True)
```

```python
# Simple show all code
pd.read_sql(
    '''
        SELECT 
            *
        FROM 
            films
    ''', 
    con
)
```

```python
# number of different characters per movie
pd.read_sql(
    '''
        SELECT 
            title,
            count(DISTINCT(characters)) AS number_characters
        FROM 
            films
        GROUP BY
            title
    ''', 
    con
)
```

```python
# number of different characters per movie
pd.read_sql(
    '''
        SELECT 
            title,
            COUNT(DISTINCT(s.name)) AS diff_species
        FROM 
            films f
            INNER JOIN people p ON p.url = f.characters
            INNER JOIN species s ON s.url = p.species
        GROUP BY
            title
        ORDER BY
            diff_species DESC
    ''', 
    con
)
```

```python
# characters with unknown species per title
pd.read_sql(
    '''
        SELECT 
            title,
            COUNT(DISTINCT(p.name)) AS unknown_species
        FROM 
            people p
            INNER JOIN films f ON f.characters = p.url
        WHERE
            species IS NULL
        GROUP BY 
            title
    ''', 
    con
)
```

```python
# characters with starships piloted > 1
pd.read_sql(
    '''
        SELECT 
            p.name,
            COUNT(DISTINCT(p.starships)) AS ss_piloted
        FROM 
            people p
            LEFT JOIN starships s ON s.url = p.starships
        GROUP BY 
            p.name
        HAVING
            ss_piloted > 1

    ''', 
    con
)
```

```python
# Planet with max AI life
pd.read_sql(
    '''
        SELECT
            planets_with_ai AS planet,
            MAX(art_life) AS art_life
        FROM (
            SELECT 
                pl.name AS planets_with_ai,
                COUNT(DISTINCT(p.name)) AS art_life
            FROM 
                people p
                INNER JOIN planets pl ON pl.url = p.homeworld
                INNER JOIN species s ON s.url = p.species       
            WHERE
                s.classification = 'artificial'
            GROUP BY
                pl.name
                )
    ''', 
    con
)
```

```python
# Vehicles and starships driven by Kenobi
pd.read_sql(
    '''
        WITH Kenobi_spaceships AS (
            SELECT 
                p.name,
                s.name AS driven
            FROM 
                people p 
                INNER JOIN starships s ON s.url = p.starships
            WHERE
                p.name = 'Obi-Wan Kenobi'
        ),
                    
        Kenobi_vehicles AS (
            SELECT 
                p.name,
                v.name AS driven
            FROM 
                people p 
                INNER JOIN vehicles v ON v.url = p.vehicles
            WHERE
                p.name = 'Obi-Wan Kenobi'
        )
        
        SELECT 
            *
        FROM
            Kenobi_spaceships 
        
        UNION
        
        SELECT 
            *
        FROM 
            Kenobi_vehicles
    ''', 
    con
)
```

```python
# Vehicles and starships driven by Sky with like
pd.read_sql(
    '''
        WITH Sky_spaceships AS (
            SELECT 
                p.name,
                s.name AS driven
            FROM 
                people p 
                INNER JOIN starships s ON s.url = p.starships
            WHERE
                p.name LIKE '%Sky%'
        ),
                    
        Sky_vehicles AS (
            SELECT 
                p.name,
                v.name AS driven
            FROM 
                people p 
                INNER JOIN vehicles v ON v.url = p.vehicles
            WHERE
                p.name LIKE '%Sky%'
        )
        
        SELECT 
            *
        FROM
            Sky_spaceships 
        
        UNION
        
        SELECT 
            *
        FROM 
            Sky_vehicles
    ''', 
    con
)
```
