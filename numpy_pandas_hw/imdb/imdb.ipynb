```python
#Split date and time from consultation date 
pd_json['date'] = pd.to_datetime(pd_json['consultation_date']).dt.date
pd_json['time'] = pd.to_datetime(pd_json['consultation_date']).dt.time
```

```python
pd_json
```

```python
# to get more insights into datetime, different columns were created for year, month, day
pd_json['year'] = pd.DatetimeIndex(pd_json['consultation_date']).year
pd_json['year'].value_counts()
pd_json['month'] = pd.DatetimeIndex(pd_json['consultation_date']).month
pd_json['month'].value_counts()
pd_json['day'] = pd.DatetimeIndex(pd_json['consultation_date']).day
pd_json['day'].value_counts()
pd_json
```

```python
# frequency of years / most common is 2022
pd_json['year'].value_counts()
```

```python
# frequency of months / most common is august, while least common is october
pd_json['month'].value_counts()
```

```python
# days people get sick, to get further info. it was decided to see the day of the week
pd_json['day'].value_counts()
```

```python
# most common days of the week / most common monday and tuesday while least common saturday and sunday
# Therefore during weekends people is magicaly healthy
pd_json['day_week'] = pd.DatetimeIndex(pd_json['consultation_date']).day_name()
pd_json['day_week'].value_counts()
```

```python
# Hours with most cases attended / as expected during the night there are less cases... therefore could be emergency only
pd_json['busy_hours'] = pd.DatetimeIndex(pd_json['consultation_date']).hour
pd_json['busy_hours'].value_counts()
```

```python
pd_json[
    (pd_json['busy_hours'] < 100)
][['cim10','busy_hours']].groupby(['cim10']).mean().sort_values(
    by=['busy_hours'],
    ascending=False)
```
