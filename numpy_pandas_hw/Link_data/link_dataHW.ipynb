```python
# Read data into json file
url = "http://app-sellaci.eu-central-1.elasticbeanstalk.com/"    
json_url = urlopen(url)
data = json.loads(json_url.read())
data
```

```python
# Changing data into their respective types
data[0][('_id')]=int(data[0][('_id')])
data[0][('doctorId')]=int(data[0][('doctorId')])
data[0][('eventId')]=int(data[0][('eventId')])
#data[0][('facturationId')]=int(data[0][('facturationId')])
data[0][('patientId')]=int(data[0][('patientId')])
data[1][('_id')]=int(data[1][('_id')])
data[1][('doctorId')]=int(data[1][('doctorId')])
data[1][('eventId')]=int(data[1][('eventId')])
#data[1][('facturationId')]=int(data[1][('facturationId')])
data[1][('patientId')]=int(data[1][('patientId')])
data
```

```python
# changing str into dict.
data[0][('patientPhoneNumber')]=eval(data[0][('patientPhoneNumber')])
data[1][('patientPhoneNumber')]=eval(data[1][('patientPhoneNumber')])
data
```

```python
#spliting cart from data and putting key inside
data_cart=[]
data_cart.append(data[0]['cart'])
data_cart[0]['_id']=(data[0]['_id'])
data_cart.append(data[1]['cart'])
data_cart[1]['_id']=(data[1]['_id'])
data_cart
```

```python
# spliting payments from carts and putting key inside
data_cart_payments=[]
data_cart_payments.append(data_cart[0]['paymentIntents'])
data_cart_payments.append(data_cart[1]['paymentIntents'])
data_cart_payments
```

```python
#creating temporary dictionarys so its possible to add id
actId1=data_cart[0]['actId']
actId2=data_cart[1]['actId']
dicttemp1 = {'actId':actId1}
dicttemp2 = {'actId':actId2}
data_cart_payments[0].append(dicttemp1)
data_cart_payments[1].append(dicttemp2)
data_cart_payments
```

```python
#Merging dictionaries inside the second payment
data_cart_payments[1]={**data_cart_payments[1][0],**data_cart_payments[1][1]}
data_cart_payments
```

```python
#taking out patient phone number from main 
data_patientPhoneNumber=[]
data_patientPhoneNumber.append(data[0]['patientPhoneNumber'])
data_patientPhoneNumber[0]['patientId']=(data[0]['patientId'])
data_patientPhoneNumber.append(data[1]['patientPhoneNumber'])
data_patientPhoneNumber[1]['patientId']=(data[1]['patientId'])
data_patientPhoneNumber
```

```python
# Same format for both
data_patientPhoneNumber[0]['dialCode'] = '+33'
data_patientPhoneNumber[0]['value']=data_patientPhoneNumber[0]['value'].replace(" ","")
data_patientPhoneNumber[0][('value')]=int(data_patientPhoneNumber[0][('value')])
data_patientPhoneNumber[1][('value')]=int(data_patientPhoneNumber[1][('value')])
data_patientPhoneNumber
```

```python
# Separate snapshot from main
data_snapshot=[]
data_snapshot.append(data[0]['snapshot'])
data_snapshot[0]['patientId']=(data[0]['patientId'])
data_snapshot.append(data[1]['snapshot'])
data_snapshot[1]['patientId']=(data[1]['patientId'])
data_snapshot
```
